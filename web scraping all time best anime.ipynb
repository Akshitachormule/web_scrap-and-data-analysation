```python
import requests

import pandas as pd

from bs4 import BeautifulSoup 

import numpy as np

from time import sleep

from random import randint

title = []
year = []
rating = []
certificate=[]
votes=[]
runtime=[]
gener=[]
voice_artist1=[]
voice_artist2=[]
voice_artist3=[]
voice_artist4=[]

pages=np.arange(1,1000,50)

for page in pages:
    page= requests.get("https://www.imdb.com/search/title/?title_type=tv_series&num_votes=1000,&genres=animation&keywords=anime&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup=BeautifulSoup(page.text,"html.parser")
    anime=soup.find_all('div',{'class':"lister-item-content"})
    sleep(randint(2,8))
    for i in anime:
        title.append(i.h3.a.text)
    
        year.append(i.find('span',{'class':'lister-item-year text-muted unbold'}).text[1:10])

        rating.append(i.find('div',{'class':'inline-block ratings-imdb-rating'}).text[2:5] if i.find('div',{'class':'inline-block ratings-imdb-rating'}) else "  ")

        x=i.p.find('span',{'class':'runtime'})
        y=x.text [0:3] if x else "0"
        runtime.append(y)

        c=i.find('span',{'class':'certificate'})
        d=c.text if c else "NA"
        certificate.append(d)

        star1= i.find('p',class_='').find_all('a')[0]
        p=star1.text if star1 else ' Na'
        voice_artist1.append(p)

        try:
            star2= i.find('p',class_='').find_all('a')[1].text if i.find('p',class_='').find_all('a')[1] else ' Na'
            voice_artist2.append(star2)
        except IndexError :
             voice_artist2.append("NA")

        try:
            star3= i.find('p',class_='').find_all('a')[2].text if i.find('p',class_='').find_all('a')[2] else ' Na'
            voice_artist3.append(star3)
        except IndexError :
             voice_artist3.append("NA")

        try:
            star4= i.find('p',class_='').find_all('a')[3].text if i.find('p',class_='').find_all('a')[3] else ' Na'
            voice_artist4.append(star4)
        except IndexError :
             voice_artist4.append("NA")



        votes.append(i.find('span',{'name':'nv'}).text)

        gener.append(i.find('span',{'class':'genre'}).text[1:])
```

```python
data=list(zip(title,year,rating,certificate,votes,runtime,gener,voice_artist1,voice_artist2,voice_artist3,voice_artist4))
data
```

```python
df=pd.DataFrame(data,columns=["title","year","rating","certificate","votes","runtime","gener","voice_artist1","voice_artist2","voice_artist3","voice_artist4"])
```
